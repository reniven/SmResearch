```python
import csv
import numpy as np
import ast
import json
import os
```

```python
k = "../Data/secondTwitterDataFull9.json"
```

```python
len(k)
```

```python
convertedArray = json.load(open(k, 'r'))
users = []
ids = []
for i, j in enumerate(convertedArray):
    ids.append(j['id_str'])
    users.append(j['user']['screen_name'])
    
for i, (j, l) in enumerate(zip(users,ids)):
    print("%d: %s" %(i,(j + "\n" + l)))
    print()

```
